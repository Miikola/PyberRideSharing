```python
# Dependencies
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
import matplotlib.lines as mlines
```

```python
# Read 1st CSV
ride_data_one = pd.read_csv("raw_data/city_data.csv")
ride_data_one.head()
```

```python
#Read 2nd CSV
ride_data_two = pd.read_csv("raw_data/ride_data.csv")
ride_data_two.head()
```

```python
# # Merge our two data frames together
combined_data_df = pd.merge(ride_data_one, ride_data_two, on="city", how='outer')
combined_data_df.head()
```

```python
sub_df = combined_data_df.loc[combined_data_df["type"]=="Suburban"]
sub_df
```

```python
organized_combined_summary_table = combined_data_df[["city","date","fare","ride_id","driver_count","type"]]
organized_combined_summary_table.sort_values("city").head()
```

```python
cityGroupBy_df = combined_data_df.groupby(["city"])
xx = cityGroupBy_df["type"]
```

```python
cityData = cityGroupBy_df.sum()
avgFarePrice = round(cityGroupBy_df["fare"].mean(),2)
riderCount = cityGroupBy_df['ride_id'].count()
totalDrivers = pd.to_numeric(cityData["driver_count"]/riderCount,downcast='signed')
the_type = xx

pyber_analysis = pd.DataFrame({
                            "City Type": the_type.max(),
                            "Average Fare ($) Per City": avgFarePrice,
                            "Total Number of Drivers Per City": totalDrivers,
                            "Total Number of Rides Per City": riderCount
                                })
pyber_summary_analysis = pyber_analysis[["City Type","Average Fare ($) Per City","Total Number of Drivers Per City","Total Number of Rides Per City"]]
pyber_summary_analysis.head()
```

# Pyber Bubble Chart 

```python
colors = {"Rural": 'gold', 'Suburban' : 'lightskyblue', 'Urban' : 'lightcoral'}
plt.figure(figsize=(10,10))
color_region = list()

for i, r in pyber_summary_analysis.iterrows():
    plt.scatter(
    x= r["Total Number of Rides Per City"],
    y = r["Average Fare ($) Per City"],
    s = r["Total Number of Drivers Per City"]*10,
    marker='o',
    facecolors=colors.get(r["City Type"][0]),
    edgecolors="black",
    alpha=0.6,
    linewidths = 1
               )

plt.title("Pyber Ride Sharing Data (2016)")
plt.xlabel("Total Number of Rides (Per City)")
plt.ylabel("Average Fare ($)")

# Set the upper and lower limits of our y axis
plt.ylim(15,45)

# Set the upper and lower limits of our x axis
plt.xlim(0,40)
plt.show()
```
